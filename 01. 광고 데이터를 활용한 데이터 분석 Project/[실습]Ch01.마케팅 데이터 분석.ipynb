### # Domain Knowledge 2 : 광고성과지표
| 광고성과지표 | = | 계산식 |
|:---------- |---------- |:---------- |
|CTR(Click Through Rate)|=|클릭수 / 노출수 * 100|
|CPM(Cost Per Mile)|=|광고비용 / 노출수 * 1000|
|CPC(Cost Per Click)|=|광고비용 / 클릭수|
|CPA(Cost Per Action)|=|광고비용 / 구매수|

```python
imp=10000 # 노출 수 Impression
clk=100 # 클릭 수 click
conv=10 # 구매 수 conversion
cost=100000 # 광고 비용 cost 10만원
```

```python
# ctr
ctr = clk/imp*100
```

```python
#ctr 출력
ctr
```

```python
# cpm
cpm = cost/imp*1000
```

```python
#cpm 출력
cpm
```

```python
# cpc
cpc = cost/clk
```

```python
#cpc 출력
cpc
```

```python
# cpa
cpa = cost/conv
```

```python
#cpa 출력
cpa
```
